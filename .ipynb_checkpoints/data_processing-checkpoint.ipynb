```python
import numpy as np
import pandas as pd

df = pd.read_csv('lastfm-dataset-360K/usersha1-artmbid-artname-plays.tsv',sep = '\t', header=None) 
```

```python
# delete data with missing values
df.dropna(subset=[2], inplace=True)
df.reset_index(drop=True, inplace=True)
display(df)

# store variables with information about the data
# info from http://ocelma.net/MusicRecommendationDataset/lastfm-360K.html
users = df[0].unique()
nUsers = len(users)
artists = df[1].unique()
nArtists = len(artists)

# add column for ranking
cur_user = 0
rank_ct = 1
rankings = [0] * len(df)
for idx, data in df.iterrows():
    #print(idx, data)
    #print(data[0])
    
    if cur_user != data[0]:
        cur_user = data[0]
        rank_ct = 1
    
    rankings[idx] = rank_ct
    rank_ct += 1

# find unique user ids, change user ids
#print(nUsers, nArtists)
user_dict = dict(zip(users.tolist(), range(nUsers)))
artist_dict = dict(zip(artists.tolist(), range(nArtists)))
```

```python
#print(rankings[0:100])

df[4] = rankings
display(df)
```
